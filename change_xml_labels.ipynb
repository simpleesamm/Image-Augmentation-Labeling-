```python
import xml.etree.ElementTree as ET
from xml.dom import minidom
import os
count = 0
path = 'C:/TensorFlow_2021/Tensorflow_OD/TFODCourse/Tensorflow/workspace/images/final_all_images/train'
new_path = 'C:/TensorFlow_2021/Tensorflow_OD/TFODCourse/Tensorflow/workspace/images/final_all_images/train'
dic = {'alphabet A':'A','alphabet B':'B','alphabet C':'C','alphabet D':'D','alphabet E':'E','alphabet F':'F','alphabet G':'G',
      'alphabet H':'H','alphabet S':'S','alphabet T':'T','alphabet U':'U','alphabet V':'V','alphabet W':'W','alphabet X':'X',
      'alphabet Y':'Y','alphabet Z':'Z','left arrow':'Left','right arrow':'Right','up arrow':'Up', 'down arrow':'Down', 
       'bullseye':'Bullseye','stop':'Stop' }
for filename in os.listdir(path):
    if not filename.endswith('.xml'): continue
    count += 1

    fullname = os.path.join(path, filename)
    tree = ET.parse(fullname)
    root = tree.getroot()
    for child in root:
        if child.tag == 'filename':
            new_filename = child.text[0:-4] + '.xml'
            #print(new_filename)
        if child.tag == 'object':
            for grand_child in child:
                if grand_child.tag == 'name':
                    if str(grand_child.text) in dic:
                        grand_child.text = dic[grand_child.text]
                        print(grand_child.text)

    new_fullname = os.path.join(new_path, new_filename)
    tree.write(new_fullname)
#         xmldoc = minidom.parse(fullname)
#         path = xmldoc.getElementsByTagName('path')
#         name = xmldoc.getElementsByTagName('name')
#         print(path, name)
print(count)
```
